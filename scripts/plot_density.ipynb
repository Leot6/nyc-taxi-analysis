```python
%matplotlib notebook
import common
import matplotlib.pyplot as plt
import numpy as np
from datetime import datetime
```

```python
df = common.load_data(nrows=1000000)
```

```python
def plot_density(n_bins=200):
    start_dt = datetime(2014, 1, 10, 5, 0)
    end_dt = datetime(2014, 1, 10, 7, 0)
    sdf = common.query_dates(df, start_dt, end_dt, "dropoff_datetime") 
    pus = common.get_pickup_geos(sdf)
    dos = common.get_dropoff_geos(sdf)
    pus_hm, xedges, yedges = np.histogram2d(pus[:, 0], pus[:, 1], bins=n_bins)
    dos_hm, xedges, yedges = np.histogram2d(dos[:, 0], dos[:, 1], bins=n_bins)
    plt.subplot(1, 2, 1)
    plt.title("Pickup density")
    plt.pcolor(pus_hm)
    plt.subplot(1, 2, 2)
    plt.title("Dropoff density")
    plt.pcolor(dos_hm)
    plt.show()
```

```python
plot_density(100)
```
